### Quick Aside: Regexes in Python

Regular expressions are often utilized in Python via the `re` library. The basic setup is to first import `re` into your environment, then create a pattern using `re.compile(r'your regex string here')`.

Notice the `r` prefixing the string. This does not stand for "regular" nor "regular expression", but "raw". Prefixing a Python string with an `r` makes it a **raw string** which does not escape the characters within. This is useful for regexes because we may want to specify our own escape characters to use in our regex. **By default, you should always use raw strings when creating regexes.**

For example, if I want to find all the instances of "cat" or "cow" (for any yoga fans) in a given bit of text, I would write something like the following code cell:

````python
import re

pattern = re.compile(r'cat|cow')
text = "I love how cats meow and cows moo. I could be a cow in another life."
pattern.findall(text)
````

The questions below rely on a different function you can call on your regex pattern, `match()`. `findall()` is good for looking for many instances among many words. `match()` is better for when the text should match your pattern in its entirety.

Using the same pattern above, matching with the string `"cat"` would return a Match object; matching with, say, `"dog"`, would return `None`. A Match object contains some useful functions which you can read about in [the documentation](https://docs.python.org/3/howto/regex.html). What we'll use here is the `group()` function, which prints out the matched text.

Warning: `match` is a recognized word in Python, so do not use it for your variable name.

````python
m = pattern.match("cat") # Do NOT use match as a variable name
m.group()
````

## Question 2 - Regular Expressions

(a) Using your list of unique tokens in the Brown corpus from the previous question, create a regex pattern and use it to print out the tokens in your list that begin with the letter `y`.

````python
import re
### YOUR CODE HERE ###
print(all_tokens_unique)

search = re.compile(r'[y]+[A-Za-z0-9._-]+')
all_tokens_string = ' '.join([str(e) for e in all_tokens_unique])
#check that string was merged properly
print(all_tokens_string)

#findall and store in search results
search_results = search.findall(all_tokens_string)


#print search results
print(search_results)
````

(b) Again using your list of unique tokens, create a regex pattern and use it to print out the tokens in your list that only contain punctuation marks; in other words, combinations of characters that are *not* alphanumeric characters.

````python
### YOUR CODE HERE ###
import re

#new regex pattern, no alphanumerics must use ^
search = re.compile(r'[^A-Za-z0-9]+')
search_results = search.findall(all_tokens_string)


#print search results
print(search_results)
````

### Question 3 - A More Complex Regex

In lecture we covered how to construct a basic email address regex. Here we will construct a street address regex. Street address formats differ between countries, so we'll work with a basic format that covers many American addresses.

The following code cell defines a Python dictionary with 5 places and their corresponding addresses. Go ahead and run the cell so the variable is available to you.

````python
addresses = {
  "University of Denver": "2199 S University Blvd, Denver, CO 80208",
  "La Chiva Colombian Restaurant": "1446 S Broadway, Denver, CO 80210",
  "Red Rocks Amphitheatre": "18300 W Alameda Pkwy, Morrison, CO 80465",
  "16th Street Mall": "1001 16th St Mall, Denver, CO 80265",
  "Telluride Ski Resort": "565 Mountain Village Blvd, Telluride, CO 81435"
}
````

Compile a regex pattern for matching street addresses such that the 5 *groups* of the address are isolated and matched. *Group* refers to capture groups, which are automatically recognized when you partition off parts of your regex using parentheses. For example,
```
r"([A-Z][a-z]+) ([A-Z][a-z]+)"
```
is a raw string that would match a first and last name, assuming there is only a capital letter at the beginning of each name. This definitely does not work for all names, highlighting that edge cases or nuanced cases are often present when it comes to formulating regexes.

Creating a Match object with the above regex on the name `"Sarah Connor"`, for example, would match two groups, the first one being `Sarah` and the second being `Connor`. The space in between is not part of either group. 

The Match object's `groups()` function will give you these isolated strings as a list.

(a) Compile a regex pattern for matching street addresses using the following criteria for each group:

1.   Street number: 1 to 5 digits
2.   Street name: 1 or more alphabetical letters or spaces
3.   City: 1 or more alphabetical letters
4.   State: 2 capital letters
5.   Zip Code: 5 digits

There should be a comma between the street name and city as well as between the city and state.

Iterate over the given dictionary of addresses and match the pattern to each address. If the pattern matches (i.e. the Match object is not `None`), print the `groups()` list to confirm you isolated each part of the address correctly.

````python
### YOUR CODE HERE ###
#
#street num: (\d{1,5}) #1 to 5 digits
#Street name: ([a-zA-Z\s]+,) 1 or more alphabetical letters
#City: ([a-zA-Z]+,) CITYNAME,
#State: ([A-Z]{2}) all caps 2 letters
#Zip: ([0-9]{5}) 5 number combo
#
#
#
#create search criteria
address_search = re.compile(r'(\d{1,5}) ([a-zA-Z\s]+,) ([a-zA-Z]+,) ([A-Z]{2}) ([0-9]{5})')
#iterate through all addresses, store in temp variable address
for address in addresses.values():
    #store matches in match
    match = address_search.match(address)
    #if match is found, print the group of matching regex
    if match is not None:

       print(match.group())


````
